```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import warnings

# Visualization settings
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
sns.set_palette("husl")
warnings.filterwarnings('ignore')

# %%
# Load the preprocessed data from Task 2
try:
    # Try loading the sentiment and theme analyzed data
    df = pd.read_csv('../data/final_reviews_with_themes.csv')
    print("Data loaded successfully. Preview:")
    print(df.head())
    
    # Extract bank name from review text (assuming bank name is mentioned in reviews)
    def extract_bank(review):
        review_lower = review.lower()
        if 'cbe' in review_lower or 'commercial bank' in review_lower:
            return 'CBE'
        elif 'boa' in review_lower or 'abyssinia' in review_lower:
            return 'BOA'
        elif 'dashen' in review_lower:
            return 'Dashen'
        else:
            return 'Unknown'
    
    df['bank'] = df['review'].apply(extract_bank)
    df = df[df['bank'] != 'Unknown']  # Remove reviews where bank couldn't be identified
    
except FileNotFoundError:
    # Sample data structure if real data isn't available
    print("Processed data file not found. Creating sample data for demonstration.")
    data = {
        'review': [
            "CBE app is great but sometimes crashes during transfers",
            "BOA app very slow loading times, needs improvement",
            "Dashen Bank has excellent user interface and easy to use",
            "I can't login to my CBE account, very frustrating",
            "Recent update made BOA app much faster"
        ],
        'sentiment_label': ['positive', 'negative', 'positive', 'negative', 'positive'],
        'sentiment_score': [0.8, -0.6, 0.9, -0.8, 0.7],
        'identified_theme': ['Performance', 'Performance', 'UI/UX', 'Login Issues', 'Performance']
    }
    df = pd.DataFrame(data)
    df['bank'] = df['review'].apply(extract_bank)

# %%
# Basic data info
print("\nData Information:")
print(df.info())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# %% [markdown]
"""
## 2. Sentiment Analysis Visualization
Let's visualize the sentiment distribution across different banks.
"""

# %%
# Sentiment distribution by bank
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='bank', hue='sentiment_label')
plt.title('Sentiment Distribution by Bank')
plt.xlabel('Bank')
plt.ylabel('Count of Reviews')
plt.legend(title='Sentiment')
plt.show()

# %%
# Average sentiment score by bank
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='bank', y='sentiment_score', estimator=np.mean)
plt.title('Average Sentiment Score by Bank')
plt.xlabel('Bank')
plt.ylabel('Average Sentiment Score')
plt.show()

# %% [markdown]
"""
## 3. Thematic Analysis Visualization
Now let's explore the key themes identified in the reviews.
"""

# %%
# Theme distribution by bank
plt.figure(figsize=(14, 7))
sns.countplot(data=df, y='identified_theme', hue='bank')
plt.title('Theme Distribution by Bank')
plt.xlabel('Count of Reviews')
plt.ylabel('Theme')
plt.legend(title='Bank')
plt.show()

# %%
# Create word clouds for each bank's reviews
def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, 
                          background_color='white', 
                          max_words=100).generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Generate word clouds for each bank
for bank in df['bank'].unique():
    bank_reviews = ' '.join(df[df['bank'] == bank]['review'].astype(str))
    generate_wordcloud(bank_reviews, f'Most Frequent Words - {bank}')

# %% [markdown]
"""
## 4. Key Insights and Recommendations
Based on our analysis, here are the key insights and recommendations for each bank.
"""

# %%
# Function to print insights in a formatted way
def print_insight(title, description, recommendation):
    print(f"\n\033[1m{title}\033[0m")
    print(f"Description: {description}")
    print(f"Recommendation: {recommendation}")

# %% [markdown]
"""
### Commercial Bank of Ethiopia (CBE) Insights
"""

# %%
# CBE Insights
print_insight(
    "Performance Issues",
    "CBE users frequently mention app crashes and performance problems.",
    "Prioritize performance optimization and stability improvements in the next update."
)

print_insight(
    "Positive Feedback",
    "Many CBE users praise the app's functionality when it works properly.",
    "Maintain current features that users appreciate while fixing performance issues."
)

# %% [markdown]
"""
### Bank of Abyssinia (BOA) Insights
"""

# %%
# BOA Insights
print_insight(
    "Speed Concerns",
    "BOA app receives consistent complaints about slow loading times.",
    "Optimize app loading speed and backend processes to improve user experience."
)

print_insight(
    "Recent Improvements Noticed",
    "Some BOA users have positively noted recent performance improvements.",
    "Continue the current optimization efforts as they're having positive impact."
)

# %% [markdown]
"""
### Dashen Bank Insights
"""

# %%
# Dashen Insights
print_insight(
    "UI/UX Appreciation",
    "Dashen Bank app receives praise for its user interface and experience.",
    "Maintain the current UI/UX standards as they're a competitive advantage."
)

print_insight(
    "Feature Requests",
    "Users request additional features like biometric authentication.",
    "Consider adding most-requested features in future updates to stay competitive."
)

# %% [markdown]
"""
## 5. Final Recommendations Summary
"""

# %%
# Final recommendations
recommendations = [
    {
        "Bank": "All",
        "Priority": "High",
        "Recommendation": "Focus on app performance and stability improvements",
        "Impact": "Would address the most common complaints across all banks"
    },
    {
        "Bank": "CBE",
        "Priority": "High",
        "Recommendation": "Fix app crashes and freezing issues",
        "Impact": "Currently the biggest pain point affecting user satisfaction"
    },
    {
        "Bank": "BOA",
        "Priority": "High",
        "Recommendation": "Optimize app loading and response times",
        "Impact": "Would significantly improve user experience"
    },
    {
        "Bank": "Dashen",
        "Priority": "Medium",
        "Recommendation": "Maintain current UI/UX quality while adding requested features",
        "Impact": "Would preserve strengths while addressing user requests"
    }
]

# Convert to DataFrame for nice display
rec_df = pd.DataFrame(recommendations)
print("\nFinal Recommendations:")
display(rec_df)

# %%
# Save visualizations to files
print("\nSaving visualizations to files...")
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='bank', hue='sentiment_label')
plt.title('Sentiment Distribution by Bank')
plt.xlabel('Bank')
plt.ylabel('Count of Reviews')
plt.legend(title='Sentiment')
plt.savefig('sentiment_by_bank.png', bbox_inches='tight', dpi=300)

plt.figure(figsize=(14, 7))
sns.countplot(data=df, y='identified_theme', hue='bank')
plt.title('Theme Distribution by Bank')
plt.xlabel('Count of Reviews')
plt.ylabel('Theme')
plt.legend(title='Bank')
plt.savefig('themes_by_bank.png', bbox_inches='tight', dpi=300)

print("Visualizations saved successfully.")
```
